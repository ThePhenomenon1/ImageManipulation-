```python
# An Image object has several useful attributes that give you basic information about the image file it was loaded from: 
# its width and height, the filename, and the graphics format (such as JPEG, GIF, or PNG). 

from PIL import Image
alien_im = Image.open('raykurz.png')
alien_im.size
```

```python
# Cropping Images (here we want the alien's face)

# Cropping an image means selecting a rectangular region inside an image and removing everything outside the rectangle. 
# The crop() method on Image objects takes a box tuple and returns an Image object representing the cropped image.
# The cropping does not happen in place  that is, the original Image object is left untouched, and the crop() method 
# returns a new Image object. 
# A boxed tuple  in this case, the cropped section  includes the left column and top row of pixels but
# only goes up to and does not include the right column and bottom row of pixels.

from PIL import Image
alien_im = Image.open('raykurz.png')
cropped_im = alien_im.crop((300, 10, 430, 220))
cropped_im.save('cropped.png')
```

```python
# This makes a new Image object for the cropped image, stores the object in cropped_im, and then calls save() on cropped_im
# to save the cropped image in cropped.png. 
```

```python
# Copying and Pasting Images onto Other Images

# The copy() method will return a new Image object with the same image as the Image object it was called on. 
# This is useful if you need to make changes to an image but also want to keep an untouched version of the original.

from PIL import Image
alien_im = Image.open('raykurz.png')
alien_copy_im = alien_im.copy()
```

```python
# Thus, if you want to tile the aliens head across the entire raykurz image, you can rename cropped.png to face_im.

face_im = alien_im.crop((300, 10, 430, 220))
face_im.size
```

```python
# Now you need a couple for loops. 

alien_im_width, alien_im_height = alien_im.size
face_im_width, face_im_height = face_im.size
alien_copy_two = alien_im.copy()
for left in range(0, alien_im_width, face_im_width):
    for top in range(0, alien_im_height, face_im_height):
        print(left, top)
        alien_copy_two.paste(face_im, (left, top))
    
alien_copy_two.save('tiled.png')
```
